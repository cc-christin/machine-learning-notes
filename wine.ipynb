```python
# import dependencies
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.datasets import load_wine
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine = load_wine()
data = wine.data
```

#### Standardize the data beforehand.

```python
data[:2]
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
wine.target[:2]
```

```python
# These target labels can optionally be used to color the t-SNE plot
labels = wine.target
```

```python
# Scale data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data)
```

#### You may have to tweak the learning_rate parameter of your t-SNE model. The normal range is 10 to 1,000.

```python
# Run t-SNE
tsne = TSNE(learning_rate=250)
tsne_features = tsne.fit_transform(scaled_features)
```

```python
# Use .shape to check dimensions that have been reduced to two 
tsne_features.shape
# (178, 2)
```

#### After reducing dimensions with t-SNE, create a scatter plot of the transformed data. How many clusters do you detect?

```python
# Plot results
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

#### In this exercise, the target labels are available, as labels = wine.target. Use these labels to color your scatter plot. 

```python
# Colored results by traget value
plt.scatter(tsne_features[:,0], tsne_features[:,1], c=labels)
plt.show()
```
